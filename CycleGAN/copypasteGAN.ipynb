```python
import os
import shutil
from pathlib import Path

# Define paths
fake_images_dir = '/content/drive/MyDrive/nordic_snow_results_desnowed_30epoch/nordic_clear2snow/test_latest/images'
labels_dir = '/content/drive/MyDrive/dataset/nvd/nvd_dataset_small/labels/test'
output_images_dir = '/content/drive/MyDrive/dataset/nvd/nvd_dataset_small/images/test_cycleGAN_30ep'
output_labels_dir = labels_dir

os.makedirs(output_images_dir, exist_ok=True)

for filename in os.listdir(fake_images_dir):
    if filename.endswith('_fake.png'):
        fake_img_path = os.path.join(fake_images_dir, filename)

        base_name = filename.replace('_fake.png', '')
        label_filename = base_name + '.txt'
        label_path = os.path.join(labels_dir, label_filename)

        new_image_path = os.path.join(output_images_dir, filename)
        new_label_path = os.path.join(output_labels_dir, filename.replace('.png', '.txt'))

        shutil.copy2(fake_img_path, new_image_path)
        print(f"Copied image: {filename} to {new_image_path}")

        if os.path.exists(label_path):
            shutil.copy2(label_path, new_label_path)
            print(f"Copied label: {label_filename} to {new_label_path}")
        else:
            print(f"Label not found for {filename} (expected {label_filename})")

print("Done!")
```
